# 6. Dropping irrelevent columns

```python
data=data.drop(['RowNumber', 'CustomerId', 'Surname'],axis=1)
```

# 7.encoding categorial data

```python
data['Geography'].unique()
```

```python
data = pd.get_dummies(data, drop_first=True, dtype=int)
```

# 8.Not handelling imbalanced

```python
data['Exited'].value_counts()
```

```python
import seaborn as sns
```

```python
x=data.drop('Exited',axis=1)
y=data['Exited']
# seperate x as the dependent data and y as the independent data
```

# 9.Handelling imbalanced data with SMOTE

```python
from imblearn.over_sampling import SMOTE 
```

```python
x_res, y_res = SMOTE () .fit_resample (x, y)
```

```python
y_res.value_counts ()
```

# 10.splitting data into traning and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split (x_res, y_res, test_size=0.20, random_state=42)
```

# 11.Feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train   
```

# 12. Random forest classifer

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```
